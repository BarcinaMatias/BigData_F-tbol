```python
import pandas as pd
import matplotlib.pyplot as plt
from mplsoccer.pitch import Pitch
import seaborn as sns
```

```python
df = pd.read_csv('messibetis5.csv')
```

```python
df['x'] = df['x']*1.2
df['y'] = df['y']*0.8
df['endX'] = df['endX']*1.2
df['endY'] = df['endY']*.8
```

```python
for x in range(len(df['x'])):
    if df['outcome'][x] == 'Successful':
        plt.plot((df['x'][x], df['endX'][x]), (df['y'][x], df['endY'][x]), color='green')
        plt.scatter(df['x'][x], df['y'][x], color='green')
    if df['outcome'][x] == 'Unsuccessful':
        plt.plot((df['x'][x], df['endX'][x]), (df['y'][x], df['endY'][x]), color='red')  # Fixed syntax error
        plt.scatter(df['x'][x], df['y'][x], color='red')

plt.xlim(0, 120)
plt.ylim(0, 80)

plt.title("Hola", color='white', size=20)
```

```python
fig ,ax = plt.subplots(figsize=(13.5,8))
fig.set_facecolor('#22312b')
ax.patch.set_facecolor('#22312b')

pitch = Pitch(pitch_type='statsbomb', pitch_color='#22312b', line_color='#c7d5cc')
fig, ax = pitch.draw(figsize=(16, 11), constrained_layout=True, tight_layout=False)
fig.set_facecolor('#22312b')

pitch.draw(ax=ax)
plt.gca().invert_yaxis()

kde = sns.kdeplot(
    x=df['x'],
    y=df['y'],
    fill=True,  # Use fill instead of shade
    shade_lowest=False,
    alpha=.5,
    n_levels=10,
    cmap='magma'
)

for x in range(len(df['x'])):
    if df['outcome'][x] == 'Successful':
        plt.plot((df['x'][x], df['endX'][x]), (df['y'][x], df['endY'][x]), color='green')
        plt.scatter(df['x'][x], df['y'][x], color='green')
    if df['outcome'][x] == 'Unsuccessful':
        plt.plot((df['x'][x], df['endX'][x]), (df['y'][x], df['endY'][x]), color='red')  # Fixed syntax error
        plt.scatter(df['x'][x], df['y'][x], color='red')

plt.xlim(0, 120)
plt.ylim(0, 80)

plt.title("Zona de influencia", color='white', size=20)
```
