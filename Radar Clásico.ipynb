```python
df = pd.read_csv('radars.csv')
```

```python
df.head()
```

```python
df['Player'] = df['Player'].str.split('\\', expand=True)[0]
```

```python
df.Player.unique()
```

```python
df.head(50)
```

```python
df = df[(df['Player'] == 'Michy Batshuayi') | (df['Player']=='Michail Antonio')].reset_index()
```

```python
df
```

```python
df = df.drop(['index', 'Rk', 'Nation', 'Pos', 'Squad', 'Age', 'Born', '90s', 'FK', 'PK', 'PKatt', 'Matches'], axis=1)
```

```python
df
```

```python
params = list(df.columns)
params = params[1:]
params
```

```python
ranges = []
a_values = []
b_values = []

for x in params:
    a = min(df[params][x])
    a = a - (a*.25)
    
    b = max(df[params][x])
    b = b + (b*.25)
    
    ranges.append((a,b))
    
for x in range(len(df['Player'])):
    if df['Player'][x] == 'Michy Batshuayi':
        a_values = df.iloc[x].values.tolist()
    if df['Player'][x] == 'Michail Antonio':
        b_values = df.iloc[x].values.tolist()
        
a_values = a_values[1:]
b_values = b_values[1:]

values = [a_values,b_values]
```

```python
title = dict (
    title_name = 'Michy Batshuayi',
    title_color = 'blue',
    subtitle_name = 'Crystal Palace',
    subtitle_color = 'blue',
    title_name_2 = 'Michail Antonio',
    title_color_2 = 'red',
    subtitle_name_2 = 'West Ham',
    subtitle_color_2 = 'red',
    title_fontsize = 18,
    subtitle_fontsize =15
)

endnote = 'futboldata_academy'
```

```python
radar = Radar ()

fig, ax= radar.plot_radar(ranges=ranges, params=params, values=values,
                         radar_color=['blue', 'red'],
                         alphas=[.75,.6], title=title, endnote=endnote,
                         compare=True)
```
